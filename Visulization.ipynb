```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({    "Month": ["Jan", "Feb", "Mar", "Apr"],

                            "Sales": [100, 150, 130, 170]})

df.plot(x="Month", y="Sales", kind="line")

plt.show()
```

```python
df.plot(x="Month", y="Sales", kind="bar")
plt.show()
```

```python
df.plot(x="Month", y="Sales", kind="barh")
plt.show()
```

```python
df2 = pd.DataFrame({    "Marks": [50, 60, 70, 80, 90, 60, 70, 85]})

df2.plot(kind="hist", bins=5)

plt.show()
```

```python
df2.plot(kind="box")
plt.show()
```

```python
df.plot(x="Month", y="Sales", kind="area")
plt.show()
```

```python
df3 = pd.DataFrame({    "Experience": [1, 2, 3, 4, 5],    "Salary": [20000, 25000, 30000, 35000, 40000]})
df3.plot(x="Experience", y="Salary", kind="scatter")
plt.show()  # Shows relationship between two numerical variables.
```

```python
df.plot(y="Sales", kind="pie", autopct="%1.1f%%") # Must select one column
plt.show()
```

```python
df2.plot(kind="density") #Smoothed version of histogram.
plt.show()
```

```python
df3.plot(kind="hexbin", x="Experience", y="Salary", gridsize=5)
plt.show()
"""Used for large scatter data.
Shows density using hexagons."""
```

```python
import matplotlib.pyplot as plt
marks = [80,60,65,95,86,99,88]
#students = ["A","B","C","D","E"]
plt.plot(marks)
plt.grid()
plt.title("Line Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(x = data,bins = 10 , color = "skyblue",edgecolor= "red")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Basic histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
marks = [80,60,65,95,86,99,88]
students = ["A","B","C","D","E","F","G"]
plt.barh(marks,students)
plt.grid()
plt.title("Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
