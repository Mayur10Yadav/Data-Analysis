```python
import pandas as pd
import matplotlib.pyplot as plt

file_path =  r"D:\Qweez Technology\Retail Store Data.csv"

data = pd.read_csv(file_path)
DF = pd.DataFrame(data)
print(DF)
DF = DF.drop_duplicates().copy()  # Removes duplicates and creates a copy

DF_grouped = DF.groupby("Region")[["Sales_Amount", "Profit"]].sum()
DF_grouped.plot(kind="bar",title= "region is generating the highest sales and profit")
plt.show()

DF_category = DF.groupby("Product_Category")["Profit"].sum()
DF_category.plot(kind="pie", autopct = "%1.1f%%", title = "Most Profitable Product category")
plt.show()


plt.scatter(DF["Discount_%"],DF["Profit"])
plt.xlabel("Discount")
plt.ylabel("profit")
plt.title("Affect of Discount on profit")
plt.grid()
plt.show()

"""#gb = DF.groupby(DF["Product_Category"])[DF["Profit"],DF["Discount_%"]]
gb = DF.groupby("Product_Category")[["Profit", "Discount_%"]].sum()
gb.plot(kind="pie",autopct="%1.1f%%")
plt.show()
gb = DF.groupby("Product_Category")["Profit"].sum()

gb.plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.show()"""
df=DF.groupby("City")["Quantity"].sum()
df.plot(kind="barh",title= " City with highest product order")
plt.show()
DF_g = DF.groupby("Product_Category")[ "Profit"].sum()
DF_g.plot(kind="bar",title = "Which product to discontinue")
plt.show()
```
